```python
import pandas as pd
import numpy as np
import requests
from pycoingecko import CoinGeckoAPI
import datetime
import cufflinks as cf
import matplotlib.pyplot as plt
import altair as alt
cg = CoinGeckoAPI()
import time
pd.set_option('display.max_rows', None)
```

```python
for year in range(14,21):
    for month in range(1,13):
        if month < 10:
            df = pd.read_html('https://coinmarketcap.com/historical/20{}0{}05/'.format(year, month), header=0, index_col='Rank')
        elif month > 9:
            df = pd.read_html('https://coinmarketcap.com/historical/20{}{}05/'.format(year, month), header=0, index_col='Rank')
        df = df[2]
        #df.drop('Unnamed: 9', axis='columns', inplace=True)
        df.to_csv('{}{}_snapshot.csv'.format(year,month))
        time.sleep(15)
        print("Done with {}-{}!".format(month, year))
```

```python
main_df = pd.DataFrame(columns=['Name', 'Symbol', 'Market Cap', 'Price', 'Circulating Supply',
       'Volume (24h)', '% 1h', '% 24h', '% 7d', 'Unnamed: 10'])
main_df
```

```python
for year in range(14,21):
    for month in range(1,13):
        #dataframes.append('{}{}_snapshot.csv'.format(year,month))
        df = pd.read_csv('{}{}_snapshot.csv'.format(year,month), index_col=None)
        df['month'] = month
        df['year'] = year
        
        main_df = pd.concat([main_df, df])

```

```python
pd.set_option('display.max_rows', 500)
pd.set_option('min_rows', 200)
```

```python
#main_df[main_df['month'] == 1]
main_df['Market Cap'] = main_df['Market Cap'].str.replace('$', '')
main_df['Market Cap'] = main_df['Market Cap'].str.replace(',', '')
main_df['Market Cap'] = pd.to_numeric(main_df['Market Cap'])
main_df.head()
```

```python
ranks = {}

for rank in range(1,11):
    means = {}
    for i in range(14,21):
        filtered = main_df[(main_df['Rank'] == rank) & (main_df['year'] == i)]
        means[i] = filtered['Market Cap'].mean()
        ranks[rank] = means
    
means
ranks
```

```python

frame = pd.DataFrame(ranks)
frame.rena
```
