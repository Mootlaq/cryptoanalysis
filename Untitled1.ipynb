```python
import pandas as pd
import numpy as np
import requests
from pycoingecko import CoinGeckoAPI
import datetime
import cufflinks as cf
import matplotlib.pyplot as plt
import altair as alt
from datetime import date
from datetime import datetime, timedelta
cg = CoinGeckoAPI()
import pandas
pd.set_option('display.max_rows', 3000)
#pd.set_option('display.max_rows', 500)
pd.set_option('min_rows', None)
```

```python
def get_history(coin):
    coin_history = cg.get_coin_market_chart_by_id(id=coin, vs_currency='usd', days='max')

    days = [i[0] for i in coin_history['prices']]
    days = pd.to_datetime(days, unit='ms')

    prices = [i[1] for i in coin_history['prices']]
#cosmos
    days = days[:-1]
    prices = prices[:-1]

#    data = 
    df = pd.DataFrame(data={'Price': prices}, index=days)
    
    return df

    
```

```python
def old_dates(df, year):
    days_inayear = 365
    total_days = year * days_inayear
    prices_list = []
    
    for i in range(len(df.index)):
        if i > (total_days - 1):
            try:
                price_then = df.loc[df.index[i] - pd.Timedelta(value=total_days, unit='d'), 'Price']
                prices_list.append(price_then)
            except:
                price_then = df.loc[df.index[i+1] - pd.Timedelta(value=total_days, unit='d'), 'Price']
                prices_list.append(price_then)
        else:
            price_then = 0
            prices_list.append(price_then)
            
    return prices_list
        

#prices_1_year_ago
```

```python
def coin_ROIs(coin):

    df = get_history(coin)

    prices_1_year_ago = old_dates(df, 1)
    prices_2_year_ago = old_dates(df, 2)
    prices_3_year_ago = old_dates(df, 3)

    df['price 1 year ago'] = prices_1_year_ago
    df['price 2 year ago'] = prices_2_year_ago
    df['price 3 year ago'] = prices_3_year_ago

    df['price 1 year ago'] = df['price 1 year ago'].replace({0.000000:np.nan})
    df['price 2 year ago'] = df['price 2 year ago'].replace({0.000000:np.nan})
    df['price 3 year ago'] = df['price 3 year ago'].replace({0.000000:np.nan})

    df['1 Year ROI'] = df['Price'] / df['price 1 year ago']
    df['2 Year ROI'] = df['Price'] / df['price 2 year ago']
    df['3 Year ROI'] = df['Price'] / df['price 3 year ago']

    df2 = df[['1 Year ROI', '2 Year ROI', '3 Year ROI']]

    ROIs = alt.Chart(df2.reset_index(), title='{} ROI'.format(coin.capitalize())).mark_line().transform_fold(['1 Year ROI', '2 Year ROI', '3 Year ROI']).encode(
        alt.X('index:T', title='Date'),
        alt.Y('value:Q', scale=alt.Scale(type='log'), title='ROI (x)'),
        color='key:N',
        tooltip='index').properties(width=820, height=600).interactive()

    return ROIs, df
```

```python
ethereum, ethdf = coin_ROIs('ethereum')
```
